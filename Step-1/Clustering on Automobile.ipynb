# Clustering

```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
```

```python
auto = pd.read_csv("New_Car_Automobile.csv")
auto.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
auto
```

```python
auto.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
auto[['car-name','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']]=auto[['car-name','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']].apply(le.fit_transform)
```

```python
auto
```

```python
from sklearn.cluster import KMeans
```

```python
x=auto[['width','height']]
```

```python
x
```

```python
km = KMeans(n_clusters=3, random_state=90)
```

```python
km.fit(x)
```

```python
identified_cluster=km.predict(x)
```

```python
d1=x.copy()
```

```python
d1['cluster'] = identified_cluster
```

```python
d1
```

# plotting 

```python
km.cluster_centers_
```

```python
vc = d1.cluster.value_counts()
```

```python
vc
```

```python
cluster_center = km.cluster_centers_
```

```python
plt.scatter(d1['width'],d1["height"],c=d1['cluster'])

plt.show()
```

```python
plt.scatter(d1['width'],d1["height"],c=d1['cluster'])

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c= 'pink')
plt.show()
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init = 'k-means++',random_state=0)
    kmeans.fit(d1)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss,c='green')
# from the above statement we can say that the optimal value of K for K-Means is 3
```
