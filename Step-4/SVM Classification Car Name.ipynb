```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('New_Car_Automobile.csv')
```

```python
data.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
data
```

```python
# low 0.0 to 0.3 (-0.3 to 0.3)
# mod -7.0 to -3.0 and 0.3 to 0.7
# high -1 to -0.7 and 0.7 to 1 
data.corr()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data['output'] = le.fit_transform(data['car-name'])
```

```python
#input
x=data.iloc[:,[7,8,9,11,14,16]]
```

```python
x
```

```python
# ouput
y = data.iloc[:,[24]]
```

```python
y
```

```python
#train ---- test
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=0)
```

```python
from sklearn.svm import SVC
```
